```python
import pandas as pd
import numpy as np
import matplotlib as matplot
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mid_year_df = pd.read_csv('Resources/FINAL_mid_year_promos_24-25.csv')
mid_year_df.head()
```

```python
year_end_df = pd.read_csv('Resources/UPDATED_end_year_promos_24-25.csv')
year_end_df.head()
```

```python
combined_df = pd.concat([mid_year_df, year_end_df]).reset_index(drop=True)
combined_df.head()
```

```python
combined_df.loc[combined_df['Employee Number'].duplicated()]
```

```python
combined_df.info()
```

```python
combined_df['New Position Start Date']=pd.to_datetime(combined_df['New Position Start Date'], format='mixed')
combined_df['Hire Date']=pd.to_datetime(combined_df['Hire Date'], format='mixed')
combined_df['Termination Date']=pd.to_datetime(combined_df['Termination Date'], format='mixed')
combined_df['Birth Date']=pd.to_datetime(combined_df['Birth Date'], format='mixed')
```

```python
combined_df.info()
```

```python
combined_df['Years Employed'] = round(
    (combined_df['New Position Start Date']-combined_df['Hire Date']).dt.days/365
,2)

combined_df['Age'] = round(
    (combined_df['New Position Start Date']-combined_df['Birth Date']).dt.days/365)

combined_df
```

```python
combined_df.to_csv('Outputs/all_promos_24-25.csv', index=False)
```

```python
median_age_esps=pd.concat([
    esp_df.groupby('Ethnicity and Race')['Age'].median(),
    esp_df.groupby('Ethnicity and Race')['Ethnicity and Race'].value_counts()
], axis=1).sort_values('Age')
median_age_esps
```

```python
fig = plt.figure(figsize=(8,4))

df_to_plot=median_age_esps

ax = sns.barplot(x='Ethnicity and Race',
            y='Age',
            data=median_age_esps, 
            palette='deep', 
            hue='Ethnicity and Race',
            errorbar=None
            )

# Display retention value on top of bars
for p in ax.patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width() / 2., height + .5, f'{height:.0f}', 
             ha='center', va='baseline', fontsize=10, color='black')

#  Display n values in center of bars
x = 0  # Counter for index tracking
for p in ax.patches: 
    num = df_to_plot['count'].values[x]  # Extract corresponding count value
    height = p.get_height()  # Get bar height
    plt.text(
        p.get_x() + p.get_width() / 2, height/2, 
        f'n={num}', ha='center', va='baseline',  # Align text
        fontsize=10, color='black'
    )
    x += 1  # Move to next index

plt.title(f"Median Age of Promoted ESP's by Ethnicity/Race {period}")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.savefig('Outputs/esps_age_by_race', bbox_inches='tight')

plt.show()
```

```python
median_employed=pd.concat([
    esp_df.groupby('Ethnicity and Race')['Years Employed'].median(),
    esp_df.groupby('Ethnicity and Race')['Ethnicity and Race'].value_counts()
], axis=1).sort_values('Years Employed')
median_employed
```

```python
fig = plt.figure(figsize=(8,4))

df_to_plot=median_employed

ax = sns.barplot(x='Ethnicity and Race',
            y='Years Employed',
            data=median_employed, 
            palette=custom_palette, 
            hue='Ethnicity and Race',
            errorbar=None
            )

for p in ax.patches[0:8]:
    height = p.get_height()
    ax.annotate(f'{height:.1f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center',
                fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

#  Display n values in center of bars
x = 0  # Counter for index tracking
for p in ax.patches: 
    num = df_to_plot['count'].values[x]  # Extract corresponding count value
    height = p.get_height()  # Get bar height
    plt.text(
        p.get_x() + p.get_width() / 2, height/2, 
        f'n={num}', ha='center', va='baseline',  # Align text
        fontsize=10, color='black'
    )
    x += 1  # Move to next index

plt.title(f'Median Years Employed of Promoted Staff by Ethnicity and Race 2023-2024')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.savefig('Outputs/years_employed_by_race_position', bbox_inches='tight')

plt.show()
```
