### Determining ending assignment for promotion year:

For staff who have a regular assignment AND temp assignment, we want to use their last regular assignment.
For staff who only have temp or regular assignments, we want to use their last assignment.

```python
duplicate_numbers=duplicates_old['Employee Number'].unique()
len(duplicate_numbers)
```

```python
# # Test code
# ending_assignments_24=pd.DataFrame()

# assignments=df_24.loc[df_24['Employee Number']==21140]
# position_types=assignments['Position Type']
# has_temp = position_types.isin(temp_types).any()
# has_non_temp = (~position_types.isin(temp_types)).any()

# if has_temp and has_non_temp:
#     print('At least one temp position')
#     assignments=assignments[~assignments['Position Type'].isin(temp_types)].sort_values('Start Date', ascending=False)
#     # ending_assignments_24=pd.concat([ending_assignments_24, assignments.iloc[0]])
#     print(assignments.iloc[0])
# # ending_assignments_24
```

```python
temp_types=['TEMPHIRE', 'TEMPCONT']
mixed_positions=[]
ending_assignments_old=pd.DataFrame()

for number in duplicate_numbers:   
    assignments=duplicates_old.loc[duplicates_old['Employee Number']==number]
    position_types=assignments['Position Type']
    has_temp = position_types.isin(temp_types).any()
    has_non_temp = (~position_types.isin(temp_types)).any()

    # If staff have temp positions and non-temp positions, keep only most recent non-temp position
    if has_temp and has_non_temp:
        mixed_positions.append(number)
        assignments=assignments[~assignments['Position Type'].isin(temp_types)].sort_values('Start Date', ascending=False)
        ending_assignments_old=pd.concat([ending_assignments_old, assignments.iloc[[0]]])

    #If staff have only temp positions or non-temp positions, keep most recent position
    else:
        assignments=assignments.sort_values('Start Date', ascending=False)
        ending_assignments_old=pd.concat([ending_assignments_old, assignments.iloc[[0]]])
        
ending_assignments_old.head()
```

```python
ending_assignments_old.info()
```

```python
# Concatenate ending assignments for duplicate staff with original dataframe (duplicates dropped) to form 
# a dataframe with ONE ending assignments for all staff

ending_assignments_old=pd.concat([
    ending_assignments_old,
    df_old
])
ending_assignments_old.info()
```

## Clean next year staff data

```python
# Load 2025 staff assignments
df_new = pd.read_csv('Resources/staff_assignments_25-26.csv')
df_new.head()
```

```python
df_new.info()
```

```python
df_new=df_new.rename(columns={'Position Type Code':'Position Type', 'Position Distributions Assignment Type Descriptions':'Position',\
          'Position Distributions Building Descriptions':'Site/Location', 'Hire Date Original':'Hire Date'})
```

```python
df_new['Position Type'].value_counts()
```

```python
# Create a new dataframe and omit summer and all temp assignments except for ITRAC Apprentice roles
# (most promotions are not going into a temp position with rare exceptions such as ITRAC Apprentices)

exclude=['EXTENDED', 'SUMMERTH', 'WSSS', 'OVERAGE', 'EMPLOAN', 'TEMP NSO', '24TEMPCO', 'COCURRIC', 'CAREER', 'TEMP CR', 'TEMPCONT', 'TEMPHIRE']
include=['ITRAC APPRENTICE']
df_new=df_new.loc\
[
    (~df_new['Position Type'].isin(exclude)) | (df_new['Position'].isin(include))
]

df_new['Position Type'].value_counts()
```

```python
# Convert Assignment Total Pay to float
df_new['Assignment Total Pay'] = df_new['Assignment Total Pay'].str.replace(',', '').astype(float)
df_new.info()
```

```python
# Drop assignments where Assignment Total Pay is zero - staff did not work in this role during this period

df_new = df_new.loc[~(df_new['Assignment Total Pay']==0)]
df_new.info()
```

```python
df_new['Start Date']=pd.to_datetime(df_new['Start Date'])
df_new['Hire Date']=pd.to_datetime(df_new['Hire Date'])
df_new.info()
```

```python
# Filter for assignments with start dates on or before cutoff date 

# Cutoff date (first day of Welcome Week of next school year)
cutoff_date='2025-08-25'

df_new=df_new[df_new['Start Date']<=cutoff_date]
df_new.info()
```

```python
# These employees have more than one assignment between 7/1 and cutoff_date. Keep latest start date with highest FTE.
df_new[df_new['Employee Number'].duplicated(keep=False)]
```

```python
# Drop employees with duplicate assignments. Keep latest start date with highest FTE.

df_new=df_new.sort_values(['Employee Number', 'Start Date', 'Entered FTE'], ascending=False)
df_new=df_new.drop_duplicates(subset='Employee Number', keep='first')
df_new.info()
```

## Find Promotions

```python
emp_numbers = df_new['Employee Number'].unique()
```

Step 2: Compare positions/position types for each employee from one year to the next to find promotions.

Loop through the data to find:

2a) Cases where an employee changed to a unique Position/Position Type combination

```python
# Step 2: Compare positions/position types for each employee from one year to the next to find promotions.

# List to hold potential promo employee numbers
possible_promos = []

# Create list to hold possible temp promos
possible_temp_promos=[]

# Loop through duplicates dataframe and identify unique role changes
for number in emp_numbers:
    position_old = ending_assignments_old.loc[ending_assignments_old['Employee Number']==number, ['Position', 'Position Type']]
    position_new = df_new.loc[df_new['Employee Number']==number, ['Position', 'Position Type']]
    if not position_old.empty and not position_new.empty:
        if tuple(position_old.iloc[0]) != tuple(position_new.iloc[0]):
            possible_promos.append(number)
                
# Number of possible promoted employees identified
print(f'Possible promos: {len(possible_promos)}')
```

```python
# Create dataframe with employees with possible promotions

possible_promos_df=pd.concat(
    [ending_assignments_old.loc[ending_assignments_old['Employee Number'].isin(possible_promos)],
     df_new.loc[df_new['Employee Number'].isin(possible_promos)]
    ])
possible_promos_df=possible_promos_df.sort_values(by=['Employee Number', 'Start Date'])
possible_promos_df.info()
```

```python
# Create Promotion column to mark identified promotions
possible_promos_df['Promotion']=np.nan
possible_promos_df['Old/New']=np.nan
possible_promos_df.info()
```

```python
# Step 4: Manually sift through possible promos to identify true promotions

# Output dataframe. Make a copy in the Resources folder and go through each employee one-by-one to identify promotions.
# Mark promotions with a 1 in Promotion column and fill in either "Old' or "New" in the Old/New column

# **Ignore previous ITRAC apprentice cohort because they will likely show up again**

possible_promos_df.to_csv('Outputs/possible_end_year_promos_24-25.csv', index=False)
```

## Load Promotion Data

```python
promos_df = pd.read_csv('Resources/UPDATED_possible_end_year_promos_24-25_cleaned.csv')
promos_df.head()
```

```python
promos_df['Old/New'].value_counts()
```

```python
old_position = promos_df.loc[promos_df['Old/New']=='Old']
old_position.head()
```

```python
old_position.rename(columns={'Position':'Old Position',\
                            'Site/Location':'Old Site/Location',\
                            'Hourly Pay':'Old Hourly Pay', 'Calendar Code':'Old Calendar Code',\
                            'Position Type':'Old Position Type'}, inplace=True)
old_position = old_position[['Employee Number', 'Employee', 'Old Position Type', 'Old Position', 'Old Site/Location', 'Old Hourly Pay', 'Old Calendar Code']]
old_position.head()
```

```python
new_position = promos_df.query('`Old/New` == "New"')
new_position.head()
```

```python
new_position.rename(columns={'Position':'New Position',\
                               'Site/Location':'New Site/Location',\
                               'Hourly Pay':'New Hourly Pay', 'Calendar Code':'New Calendar Code',\
                            'Start Date':'New Position Start Date',
                            'Position Type':'New Position Type'}, inplace=True)
new_position = new_position[['Employee Number', 'Employee', 'New Position Type', 'New Position', 'New Site/Location', 'New Hourly Pay', 'New Calendar Code',\
                             'New Position Start Date', 'Hire Date', 'Termination Date', 'Ethnicity and Race', 'Gender', 'Birth Date']]
new_position.head()
```

```python
end_year_promos = pd.merge(old_position, new_position, how='left', on=['Employee Number', 'Employee'])
end_year_promos.head()
```
