```python
import pandas as pd
import numpy as np
import matplotlib as matplot
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
# Import active and inactive students

student_df=pd.read_csv('Resources/student_data_042425.csv')
student_df.head()
```

```python
student_df.info()
```

```python
# Rename columns
student_df=student_df.rename(columns={'Perm ID':'District287 Id', 'VerboseAge':'Student Age', 'Organization Abbr Name':'Site/Program',\
                                     'Enter Date':'Classroom Enter Date', 'Leave Date':'Classroom Leave Date', 'Staff ID':'Employee Number',\
                                     'Gender':'Student Gender', 'Resolved Race/Ethnicity':'Student Race'})
```

```python
student_df.info()
```

```python
student_df[student_df['District287 Id'].duplicated()]
```

```python
# Convert enter date to datetime
student_df['Original Enter Date']=pd.to_datetime(student_df['Original Enter Date'])
```

```python
# Calculate years enrolled for students
current_date='2024-12-04'
student_df['Years Enrolled']=round(((pd.to_datetime(current_date)-student_df['Original Enter Date']).dt.days/365),2)
student_df.head()
```

```python
# Convert student age to integer dtype
student_df['Student Age_Years']=student_df['Student Age'].str.split(expand=True)[0].astype('int')
student_df.head()
```

```python
student_df.info()
```

```python
# Get program counts
student_df['Site/Program'].value_counts()
```

```python
# Code programs into schools

nec_sped=['302N NEC Tier 2', '303N NEC Tier 3']
wec_sped=['302W WEC Tier Two', '302WT WEC Tier 2 Trn', '303W WEC Tier 3', '303WT WEC Tier 3 Trn']
sec_sped=['302S SEC Tier 2', '303S SEC Tier 3', '303ST SEC Tier 3 Trn']
ctc=['301CT CTC Tier 1 Trn', 'CTC Pathways EPC 32', 'Gateway Academy 801']
abec=['302AT ABEC Tier 2 Tr', '302A ABEC Tier 2', '303AT ABEC Tier 3 Tr', '303A ABEC Tier 3']

student_df.loc[student_df['Site/Program'].isin(nec_sped), 'School']='NEC SPED'
student_df.loc[student_df['Site/Program'].isin(wec_sped), 'School']='WEC SPED'
student_df.loc[student_df['Site/Program'].isin(sec_sped), 'School']='SEC SPED'
student_df.loc[student_df['Site/Program'].isin(abec), 'School']='ABEC'
student_df.loc[student_df['Site/Program']=='NECA 772', 'School']='NECA'
student_df.loc[student_df['Site/Program']=='SECA 622', 'School']='SECA'
student_df.loc[student_df['Site/Program']=='W-Alt 800', 'School']='W-ALT'
student_df.loc[student_df['Site/Program']=='301CT CTC Tier 1 Trn', 'School']='CTC Transition'
student_df.loc[student_df['Site/Program']=='Gateway Academy 801', 'School']='Gateway Academy'
student_df.loc[student_df['Site/Program']=='CTC Pathways EPC 32', 'School']='CTC Pathways'
student_df.info()
```

```python
# Check uncoded programs
student_df[student_df['School'].isna()]['Site/Program'].value_counts()
```
